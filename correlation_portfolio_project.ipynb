```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)


df=pd.read_csv("E:\Analytics\python\movies.csv")
df.head(10)
```

```python
# mising data

for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,missing))
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# detect datatypes

df.dtypes
```

```python
# changing data types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#checking the released date and year 

df['year_check'] = df['released'].str.extract(pat = '([0-9]{4})').astype('int64')

 # pat is a regular expression pattern
 # 0-9 is used to check the string contains the number 0 to 9 and we cannot reduce the numbers like 0-7
 # {4} retrive the four consequtive numbers, we can change according to our use
```

```python
df.head()
```

```python
# sorting values but not changing on orginal area

df = df.sort_values(by=['gross'],inplace=False,ascending=False)
df
```
